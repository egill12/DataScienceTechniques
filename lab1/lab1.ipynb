```python
''''
Lab 1:
 Iris dataset
'''

import pandas as pd
from matplotlib import pyplot as plt

file_name = r"C:\Users\edgil\Documents\Masters\DSTA\code\iris.csv"
data = pd.read_csv(file_name)
print(data.columns)

print(data.head())
#plt.plot(data['PetalWidthCm'],data['SepalWidthCm'])

species = set(data['Species'])
mean_values = pd.DataFrame()
var_values = pd.DataFrame()

for plant in species:
     mean_values[plant] = data[data['Species']== plant].mean()
     var_values[plant] = data[data['Species']== plant].var()

setosa = data[data['Species'] == "Iris-setosa"]
versicolor = data[data['Species'] == "Iris-versicolor"]
virginica = data[data['Species'] == "Iris-virginica"]
target = data.Species
```

```python
# plotting the charts in a nice way
plt.figure()
# create a chart with 1 row and 2 columns
fig, ax = plt.subplots(2,2,figsize=(21, 10))
#ax[0,0] means place the chart in the 0,0 area of  4 chart plot, 1,1 is bottom rightHS
setosa.plot(x="SepalLengthCm", 
            y="SepalWidthCm", kind="scatter",ax=ax[0,0],label='setosa',color='r')
versicolor.plot(x = "SepalLengthCm", 
                y = "SepalWidthCm", kind = "scatter", ax = ax[0,0], label = "versi",
                color = "g" )
virginica.plot(x = "SepalLengthCm", 
                y = "SepalWidthCm", kind = "scatter", ax = ax[0,0], label = "virgin",
                color = "b" )
```

```python
# run cluster algorigthm on the data
from sklearn.cluster import KMeans
train_data = data.loc[:,data.columns != "Species"].iloc[:,1:]
model = KMeans(3,"random").fit(train_data)
train_data.head()
```

```python
est = model.labels_
output = model.predict(train_data)
#model.predict(virginica.loc[:,setosa.columns != "Species"])
# plotting the charts in a nice way
plt.figure()
# create a chart with 1 row and 2 columns
fig, ax = plt.subplots(1,2,figsize=(21, 10))
plt.scatter(train_data.iloc[:,1], train_data.iloc[:,2],train_data.iloc[:,3] ,c = output/2)
#plt.scatter(setosa, train_data[:,2], train_data[:,3], color='r')
#train_data.iloc[:,1]
```
